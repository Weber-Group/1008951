```python
import psana
import numpy as np

##########################
runNumber = 13
experiment = 'cxi100895124'
##########################

dsname = f'exp={experiment}:run={runNumber}:smd'
ds = psana.DataSource(dsname)
det = psana.Detector('alvium_u1')

# Collect 1000 lineouts from events after event 1000
n_images = 20
img = np.zeros((n_images, 1000, 1000))

processed = 0
for nevt, evt in enumerate(ds.events()):
    if nevt <= 0:
        continue  # skip first 1000 events

    if processed >= n_images:
        break

    img[processed] = det.image(evt)
    if img[processed] is None:
        continue  # skip if image is missing

    # try:
    #     lineout[processed] = np.sum(img, axis=0)
    processed += 1
    # except Exception as e:
    #     print(f"Skipping event {nevt} due to error: {e}")

print(f"Collected {processed} lineouts.")

        
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
plt.figure(figsize=[17,17])
for i in range(0,9):
    plt.subplot(3,3,i+1)
    plt.pcolor(img[i+9],norm=LogNorm())
    plt.title(f'Run {runNumber}, Shot: {i+9}')
```
